```python
import csv

def main():
    f=open('q4.csv',encoding='cp949')
    data=csv.reader(f)
    header=next(data)
    
    B1=-99999999999
    L1=99999999999
    B2=-999999999999
    L2=999999999999999
    B3=-9999999999999
    L3=9999999999999
    B4=-9999999999999
    L4=99999999999999
    line1_1=""
    line1_2=""
    line2_1=""
    line2_2=""
    line3_1=""
    line3_2=""
    line4_1=""
    line4_2=""
    for row in data:
        if row[1]=="1호선" or row[1]=="2호선" or row[1]=="3호선" or row[1]=="4호선":
            row[-2]=int(row[-2])
            row[-3]=int(row[-3])
            if(row[1]=="1호선"):
                if(row[-3]+row[-2]>B1):
                    B1=row[-3]+row[-2]
                    line1_1=row[3]
                if(row[-3]+row[-2]<L1):
                    L1=row[-3]+row[-2]
                    line1_2=row[3]
                    
            elif row[1]=="2호선":
                if(row[-3]+row[-2]>B2):
                    B2=row[-3]+row[-2]
                    line2_1=row[3]
                if(row[-3]+row[-2]<L2):
                    L2=row[-3]+row[-2]
                    line2_2=row[3]
                    
            elif row[1]=="3호선":
                if(row[-3]+row[-2]>B3):
                    B3=row[-3]+row[-2]
                    line3_1=row[3]
                if(row[-3]+row[-2]<L3):
                    L3=row[-3]+row[-2]
                    line3_2=row[3]
                    
            elif row[1]=="4호선":
                if(row[-3]+row[-2]>B4):
                    B4=row[-3]+row[-2]
                    line4_1=row[3]
                if(row[-3]+row[-2]<L4):
                    L4=row[-3]+row[-2]
                    line4_2=row[3]
                
    
    f.close()
    print("***Subway Report for Seoul on March 2023***")
    print("Line1:")
    print("Busiest Station:{}({})".format(line1_1,B1))
    print("Least used Station:{}({})".format(line1_2,L1))
    print("Line2:")
    print("Busiest Station:{}({})".format(line2_1,B2))
    print("Least used Station:{}({})".format(line2_2,L2))
    print("Line3:")
    print("Busiest Station:{}({})".format(line3_1,B3))
    print("Least used Station:{}({})".format(line3_2,L3))
    print("Line4:")
    print("Busiest Station:{}({})".format(line4_1,B4))
    print("Least used Station:{}({})".format(line4_2,L4))
    
if __name__=="__main__":
    main()
```
